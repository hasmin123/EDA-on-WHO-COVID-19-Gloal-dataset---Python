# Exploratory Data Analysis

### Deaths - cumulative total by top 10 countries

```python
x=df.nlargest(10,'Deaths - cumulative total')['Country']
y=df.nlargest(10,'Deaths - cumulative total')['Deaths - cumulative total']
plt.bar(x,y,color=c,width=0.5,align="center",linewidth=3,linestyle=":",alpha=0.8)
plt.xticks(rotation=90)
plt.title("Deaths - cumulative total by top 10 countries",fontsize=15)
plt.xlabel("Country",fontsize=10)
plt.ylabel("Deaths - cumulative total",fontsize=10)
plt.show()
```

### Cases - cumulative total by WHO Region

```python
ax = df.groupby(['WHO Region'], as_index=False)['Cases - cumulative total'].sum().sort_values(by='Cases - cumulative total', ascending=False)
sns.barplot(x = 'WHO Region',y= 'Cases - cumulative total',hue='WHO Region',data =ax)
plt.title("Cumulative total cases by WHO Region",fontsize=14)
plt.xticks(rotation=90)
plt.show()
```

### Cumulative total death by WHO Region

```python
ax = df.groupby(['WHO Region'], as_index=False)['Deaths - cumulative total'].sum().sort_values(by='Deaths - cumulative total', ascending=False)
sns.barplot(x = 'WHO Region',y= 'Deaths - cumulative total',hue='WHO Region',data =ax)
plt.title("Cumulative total death by WHO Region",fontsize=14)
plt.xticks(rotation=90)
plt.show()
```

## Death Percentage by WHO Region

```python
ax = df.groupby(['WHO Region'], as_index=False)['Deaths - cumulative total'].sum().sort_values(by='Deaths - cumulative total', ascending=False)
x=ax["Deaths - cumulative total"]
y=ax["WHO Region"]

c=["y","m","g","r","b","orange"]

plt.pie(x,labels=y,colors=c,autopct="%0.1f%%",shadow=True,wedgeprops=({"linewidth":1,"edgecolor":"b"}),radius=1,labeldistance=1.1,startangle=90,textprops={"fontsize":10},counterclock=False)

plt.legend(loc=4)
plt.show()
```
